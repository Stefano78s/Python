```python
covid_data["continent"].isnull().sum()
```

```python
# ho trasformato nel mentre la data in quanto object da info ed utile per le analisi future
covid_data["date"] = pd.to_datetime(covid_data["date"])
```

```python
elementi_null = covid_data[covid_data["continent"].isnull()][["continent","location"]]
elementi_null["location"].unique()
```

```python
# vado a sostituire i valori mancanti di continet come null tramite il ciclo for
continents = ["Africa","Asia","Europe","North America","Oceania","South America"]

for cont in continents:
    filtro_località = covid_data["location"] == cont
    filtro_null = covid_data["continent"].isnull()
    filtro_replace = filtro_località & filtro_null
    covid_data.loc[filtro_replace,"continent"]= cont
```

```python
# verifica null con diminuzione rispetto alla riga 9
covid_data["continent"].isnull().sum()
```

```python
#ho creato una copia del dataset priva di elementi null
covid_data_new = covid_data[covid_data["continent"].notnull()]
```

```python
covid_data_new.info()
```

```python
# Calcola il massimo numero di casi totali per ogni continente
totali_casi = covid_data_new.groupby("continent")["total_cases"].max()
totali_casi
```

```python
# Seleziono i dati relativi a "World" e trova il massimo valore di casi totali registrati
filto_World = covid_data["location"] == "World"
casi_tot_World = covid_data.loc[filto_World,"total_cases"].max()
casi_tot_World
```

```python
# Calcolo la percentuale dei casi totali di ogni continente rispetto al totale mondiale, arrotondata a due decimali
percentuale = (totali_casi/casi_tot_World)*100
percentuale.round(2)
```

```python
# Estraggo l'anno dalla colonna 'date' e creo una nuova colonna 'anno'
covid_data_new["anno"] = covid_data_new["date"].dt.year
covid_data_new["anno"]
```

```python
# Filtro i dati per l'Italia, per l'anno 2022 e per i giorni con nuovi casi diversi da zero
filtro_italia = covid_data_new["location"] == "Italy"
filtro_anno = covid_data_new["anno"] == 2022
filtro_new_cases = covid_data_new["new_cases"] != 0
covid_data_italia22 = covid_data_new.loc[filtro_italia & filtro_anno & filtro_new_cases]
covid_data_italia22
```

```python
# Importo la libreria matplotlib.pyplot per creare grafici/report utili alla causa
import matplotlib.pyplot as plt
```

```python
# Creo un grafico a linee che mostra l'andamento dei casi totali in Italia nel 2022 con linea blue
covid_data_italia22.plot(
    x="date", 
    y="total_cases", 
    color="blue", 
    title="Casi totali in Italia nel 2022"
)
plt.xlabel("Data")
plt.ylabel("Numero di casi")
plt.show()
```

```python
# Creo un grafico a linee che mostra l'andamento dei nuovi casi in Italia nel 2022 con linea rossa
covid_data_italia22.plot(x="date", y="new_cases", color="red", title="Nuovi casi Italia 2022")
plt.xlabel("Data")
plt.ylabel("Nuovi casi")
plt.show()
```

```python
# Estraggo il mese dalla colonna 'date' e creo una nuova colonna 'mese' con questi valori
covid_data_new["mese"] = covid_data_new["date"].dt.month
covid_data_new["mese"]
```

```python
# Importo la libreria seaborn per creare grafici statistici più avanzati
import seaborn as sns
```

```python
# Seleziono i dati di Italia, Germania e Francia da maggio 2022 ad aprile 2023 per analisi con boxplot
filtro_Italy = covid_data_new["location"] == "Italy"
filtro_Germany = covid_data_new["location"] == "Germany"
filtro_France = covid_data_new["location"] == "France"
filtro_unico = filtro_Italy | filtro_Germany | filtro_France
filtro_data1 = (covid_data_new["mese"] >= 5) & (covid_data_new["anno"] == 2022)
filtro_data2 = (covid_data_new["mese"] <= 4) & (covid_data_new["anno"] == 2023)
filtro_tempo = filtro_data1 | filtro_data2
nazioni_boxplot = covid_data_new.loc[filtro_unico & filtro_tempo]
nazioni_boxplot
```

```python
# Creo un boxplot per confrontare il numero di pazienti in terapia intensiva tra Italia, Germania e Francia
sns.boxplot(data=nazioni_boxplot, x="location", y="icu_patients")
plt.title("Pazienti in terapia intensiva")
plt.show()
```

```python
# Filtro i dati del 2021 per Italia, Germania, Francia e Spagna per un'analisi comparativa
filtro_Italy = covid_data_new["location"] == "Italy"
filtro_Germany = covid_data_new["location"] == "Germany"
filtro_France = covid_data_new["location"] == "France"
filtro_Spain = covid_data_new["location"] == "Spain"
filtro_unico = filtro_Italy | filtro_Germany | filtro_France | filtro_Spain
filtro_2021 = covid_data_new["anno"] == 2021
nazioni_boxplot2 = covid_data_new.loc[filtro_unico & filtro_2021]
nazioni_boxplot2
```

```python
# Calcolo la somma totale dei pazienti ospedalizzati nel 2021 per ciascuna delle quattro nazioni selezionate grazie al groupby e la funzione sum
nazioni_boxplot2.groupby("location")["hosp_patients"].sum()
```

```python
# Creo un grafico a barre con la somma dei pazienti ospedalizzati nel 2021 per ogni nazione selezionata
# Nel dataset relativo alla Germania per l’anno 2021, i valori della colonna "hosp_patients" risultano essere mancanti.
# Quindi, è meglio considerare questi dati mancanti come tali e non provare a riempirli con stime, per non rischiare di avere risultati sbagliati nell’analisi.
nazioni_boxplot2.groupby("location")["hosp_patients"].sum().plot(kind="bar")
plt.title("Pazienti ospedalizzati nel 2021 per nazione")
plt.xlabel("Nazione")
plt.ylabel("Totale pazienti ospedalizzati")
plt.show()
```
